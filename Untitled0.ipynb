## Identify High-Performance Outliers

### Subtask:
Filter and analyze the products that demonstrate high sales volume despite low inventory levels to suggest high turnover rates.


**Reasoning**:
I will filter the DataFrame to identify high-performance outliers based on the 95th percentile of 30-day sales and low inventory levels, calculate their turnover ratio, and display the top 10 examples.



```python
import numpy as np

# 1. Define high-performance criteria (95th percentile for 30-day sales)
sales_threshold = df['30天销量'].quantile(0.95)
inventory_mean = df['库存'].mean()

print(f"Sales Threshold (95th percentile): {sales_threshold:.2f}")
print(f"Inventory Threshold (Mean): {inventory_mean:.2f}")

# 2. Filter for high sales and low inventory
outliers = df[(df['30天销量'] > sales_threshold) & (df['库存'] < inventory_mean)].copy()

# 3. Calculate turnover ratio (Sales / Inventory)
# Using Inventory + 1 to avoid division by zero if any inventory is 0
outliers['Turnover_Ratio'] = outliers['30天销量'] / (outliers['库存'] + 1)

# 4. Display top 10 outliers by Turnover Ratio
print(f"\nIdentified {len(outliers)} high-turnover outliers.")
top_outliers = outliers[['产品名称', '30天销量', '库存', 'Turnover_Ratio']].sort_values(by='Turnover_Ratio', ascending=False).head(10)

display(top_outliers)
```

## Correlation and Business Insights

### Subtask:
Analyze the correlation between inventory and sales to derive business insights regarding the low-stock/high-turnover model.


**Reasoning**:
I will calculate the correlation, visualize the turnover ratio distribution, and compare pricing strategies to derive business insights for the high-turnover model.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculate Pearson correlation
correlation = df[['库存', '30天销量']].corr().iloc[0, 1]
print(f"Pearson Correlation between Inventory and 30-Day Sales: {correlation:.4f}")

# 2. Visualize Turnover_Ratio distribution for outliers
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(outliers['Turnover_Ratio'], bins=20, kde=True, color='purple')
plt.title('Distribution of Turnover Ratio (High-Performers)')
plt.xlabel('Turnover Ratio (Sales/Inventory)')

plt.subplot(1, 2, 2)
sns.boxplot(x=outliers['Turnover_Ratio'], color='orchid')
plt.title('Boxplot of Turnover Ratio')

plt.tight_layout()
plt.show()

# 3. Compare average '卖家价' (Seller Price)
avg_price_outliers = outliers['卖家价'].mean()
avg_price_rest = df[~df.index.isin(outliers.index)]['卖家价'].mean()

print(f"Average Seller Price (High-Turnover Outliers): {avg_price_outliers:.2f}")
print(f"Average Seller Price (Rest of Dataset): {avg_price_rest:.2f}")

# 4. Print Business Insights Summary
print('\n--- Business Implications ---')
print(f"The low correlation ({correlation:.4f}) suggests that high inventory levels do not directly translate to higher sales.")
print("The data supports a 'Just-in-Time' (JIT) model where high-performing items maintain extremely lean stock relative to their velocity.")
if avg_price_outliers < avg_price_rest:
    print("High-turnover items appear to have lower average pricing, suggesting a volume-driven strategy.")
else:
    print("High-turnover items maintain premium or standard pricing, suggesting high demand regardless of price.")
```

```python
import pandas as pd

# 1. Group by '等级' and calculate mean and median using .agg()
# 2. Rename columns for clarity
service_rate_agg = df.groupby('等级')['30天服务率'].agg(['mean', 'median']).reset_index()
service_rate_agg.rename(columns={'mean': 'Mean Service Rate', 'median': 'Median Service Rate'}, inplace=True)

# 3. Sort by Mean Service Rate in descending order
service_rate_agg_sorted = service_rate_agg.sort_values(by='Mean Service Rate', ascending=False)

# 4. Display the final aggregation table
print("Aggregated 30-Day Service Rate by Product Grade:")
display(service_rate_agg_sorted)

# Store result in variable for later use if needed
grade_trends = service_rate_agg_sorted
```
